## Data types 
- Numeric types: Integers, floating-point numbers, and complex numbers
- Boolean type: True or False
- Sequence types: Strings, lists, and tuples
- Mapping type: Dictionaries

## Expressions and Variables
**Expression** is a combination of values, variables, operators, and function calls that evaluates to a single value.

For example, the following code is 4 + 4 is an expression that evaluates to the value 8.
```
4 + 4
```

**Variable** is a named location in memory where a value can be stored. 

For example, the following code creates a variable MyName and assigns it the value 'Hanadi' 
```
MyName = 'Hanadi'
```

#Single operations
**Single operations** are simple expressions that involve only one value or variable and an operator. The operator performs a specific operation on the value or variable, and the resulting value is returned as the output of the expression.
- Absolute value : abs()
- Length of a string : len()
- converting to string : str()


#Lists and Tuples
**A list** is used to store multiple items in one variable and can be created using *square brackets.* 


```
MyList= []
```
**A tuples** also can store multiple items in a single variable and can be declared using *parentheses.*


```
MyTuple=()
```

#Dictionaries
**A dictionary** is a collection of key-value pairs. The dictionary keys must be unique.


```
MyDic= {'Key': 'Value'}
```

````python
students_name = ['Hanadi', 'Sara', 'Nora', 'Hana', 'Amal'] # List include name of Students
course_name = ('Math', 'English', 'Big data')   #Tuple include name of Coures
grade_info = {'5': 'Excellent', '4': 'Very Good', '3': 'Good', '2': 'Satisfactory', '1': 'Pass', '0': 'Fail'}   #Dictionary inclides are keys and values
````
